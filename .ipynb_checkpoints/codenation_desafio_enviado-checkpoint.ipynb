# Desafio Codenation - Previsão das Notas de Matemática do ENEM 2016 utilizando Machine Lerning

### Importando as bibliotecas necessárias

```python
import pandas as pd
import numpy as np
from sklearn import metrics
import matplotlib.pyplot as plt
import seaborn as sns
```

### Carregando as bases de dados para o data frame

```python
path_train = 'D:\Codenation\\train.csv'
path_test = 'D:\Codenation\\test.csv'

df_enem_train = pd.read_csv(path_train)
df_enem_test = pd.read_csv(path_test)
df_enem_train.head()
```

### Verificando quais features possuem maior corretação com a váriavel objetivo da predição

```python
df_enem_train.corr()
```

### Seleção das features para a criação do modelo de predição

```python
# A seleção de features é um processo pessoal onde você deve analisar correlação entre as váriaveis e testar na prática 
# quais outras variaveis podem impactar o resultado do seu modelo.

# Selecionei todas variaveis NOTA e as variaveis que indicam o ano de conclusão do Ensino Médio, tipo de escola e
# se a pessoa é treineiro ou não.

features = [
            'NU_NOTA_CN',
            'NU_NOTA_CH',
            'NU_NOTA_LC',
            'NU_NOTA_REDACAO',
            'NU_NOTA_COMP1',
            'NU_NOTA_COMP2',
            'NU_NOTA_COMP3',
            'NU_NOTA_COMP4',
            'NU_NOTA_COMP5',
            'TP_ANO_CONCLUIU',
            'TP_ESCOLA',
            'IN_TREINEIRO']

features_corr = [
                'NU_NOTA_MT',
                'NU_NOTA_CN',
                'NU_NOTA_CH',
                'NU_NOTA_LC',
                'NU_NOTA_REDACAO',
                'NU_NOTA_COMP1',
                'NU_NOTA_COMP2',
                'NU_NOTA_COMP3',
                'NU_NOTA_COMP4',
                'NU_NOTA_COMP5',
                'TP_ANO_CONCLUIU',
                'TP_ESCOLA',
                'IN_TREINEIRO']
```

### Verificando se existem valores nulos na base de dados

```python
df_enem_train[features].isnull().sum()
```

### Preenchendo os dados nulos com zero e verificando a distribuição das notas em um gráfico

```python
# Podemos ver no gráfico que ficamos com muitos registros com valores igual a zero, precisaremos tratar os dados para
# utiliza-los na predição

x0 = df_enem_train['NU_NOTA_CN'].fillna(0)
x1 = df_enem_test['NU_NOTA_CN'].fillna(0)

sns.distplot(x0)
sns.distplot(x1)
plt.legend(labels=['TRAIN','TEST'], ncol=2, loc='upper left')
```

### Selecionamos para nossa base de dados somente os registros com valores diferente de 0 e Nulo

```python
# Após selecionamos no data frame somente os valores que possuem notas diferente de 0 e Nulo o gráfico exibe uma
# distribuição completamente diferente da anterior

df_enem_train = df_enem_train.loc[
                (df_enem_train['NU_NOTA_CN'].notnull()) & (df_enem_train['NU_NOTA_CN'] != 0) &
                (df_enem_train['NU_NOTA_CH'].notnull()) & (df_enem_train['NU_NOTA_CH'] != 0) &
                (df_enem_train['NU_NOTA_LC'].notnull()) & (df_enem_train['NU_NOTA_LC'] != 0) &
                (df_enem_train['NU_NOTA_REDACAO'].notnull()) & (df_enem_train['NU_NOTA_REDACAO'] != 0)]

df_enem_test = df_enem_test.loc[
                (df_enem_test['NU_NOTA_CN'].notnull()) & (df_enem_test['NU_NOTA_CN'] != 0) &
                (df_enem_test['NU_NOTA_CH'].notnull()) & (df_enem_test['NU_NOTA_CH'] != 0) &
                (df_enem_test['NU_NOTA_LC'].notnull()) & (df_enem_test['NU_NOTA_LC'] != 0) &
                (df_enem_test['NU_NOTA_REDACAO'].notnull()) & (df_enem_test['NU_NOTA_REDACAO'] != 0)]

x0 = df_enem_train['NU_NOTA_CN'].fillna(0)
x1 = df_enem_test['NU_NOTA_CN'].fillna(0)

sns.distplot(x0)
sns.distplot(x1)
plt.legend(labels=['TRAIN','TEST'], ncol=2, loc='upper left')
```

### Verificando se ainda possuimos valores nulos no data frame

```python
df_enem_test[features].isnull().sum()
```

```python
df_enem_test['NU_NOTA_COMP1'].fillna(0, inplace=True)
df_enem_test['NU_NOTA_COMP2'].fillna(0, inplace=True)
df_enem_test['NU_NOTA_COMP3'].fillna(0, inplace=True)
df_enem_test['NU_NOTA_COMP4'].fillna(0, inplace=True)
df_enem_test['NU_NOTA_COMP5'].fillna(0, inplace=True)

df_enem_train['NU_NOTA_COMP1'].fillna(0, inplace=True)
df_enem_train['NU_NOTA_COMP2'].fillna(0, inplace=True)
df_enem_train['NU_NOTA_COMP3'].fillna(0, inplace=True)
df_enem_train['NU_NOTA_COMP4'].fillna(0, inplace=True)
df_enem_train['NU_NOTA_COMP5'].fillna(0, inplace=True)
```

### Utilizamos o Random Forest para a previsão das notas e exibimos algumas métricas de erro

```python
y_train = df_enem_train['NU_NOTA_MT']
X_train = df_enem_train[features]
X_test = df_enem_test[features]

from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)

from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(
            criterion='mae',
            max_depth=8,
            max_leaf_nodes=None,
            min_impurity_split=None,
            min_samples_leaf=1,
            min_samples_split=2,
            min_weight_fraction_leaf=0.0,
            n_estimators=500,
            n_jobs=-1,
            random_state=0,
            verbose=0,
            warm_start=False)

regressor.fit(X_train, y_train)

y_pred_test = regressor.predict(X_test)
y_pred_train = regressor.predict(X_train)

print('MAE: ', metrics.mean_absolute_error(y_train,y_pred_train).round(8))
print('MSE: ', metrics.mean_squared_error(y_train, y_pred_train).round(8))
print('RMSE: ', np.sqrt(metrics.mean_squared_error(y_train, y_pred_train).round(8)))
```

### Verifica o score dos dados e a previsão

```python
regressor.score(X_train, y_train)
```

### Guarda o resultado em um data frame contendo a inscrição do Aluno e a nota de Matemática que foi prevista

```python
df_enem_test['NU_NOTA_MT'] = y_pred_test
df_enem_test.columns
df_resultado = df_enem_test[['NU_INSCRICAO','NU_NOTA_MT']]
df_resultado.set_index('NU_INSCRICAO', inplace=True)
df_resultado.head()
```

### Exportando a variavel que era o objetivo da predição

```python
df_resultado.to_csv('D:\\aswer.csv')
```
